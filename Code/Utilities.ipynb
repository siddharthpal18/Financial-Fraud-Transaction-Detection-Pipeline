```python
from pyspark.sql.types import StructType

def read_csv_df(
    path : str,
    header: bool = True,
    infer_schema: bool = False,
    schema: StructType | str | None = None,
    delimiter: str = ","
):
    
    reader = spark.read.options(
        header = header,
        inferSchema = infer_schema,
        delimiter = delimiter,
        
    )

    if schema:
        reader = reader.schema(schema)
    
    df = reader.csv(path)
    return df
```
